```python
%matplotlib inline
import requests
import json 
import matplotlib.pyplot as plt
import pandas
from datetime import datetime
import seaborn as sns
import glob
from os.path import basename
sns.set_context('poster')
import re
```

```python
speeches = glob.glob('../results/receptiviti/trump-speeches/*')
```

```python
dataframes = []
for s in speeches:
    j = json.load(open(s))
    fields = {k:v for k,v in j['liwc_scores'].iteritems() if k != 'categories'}
    fields.update(j['liwc_scores']['categories'])
    df = pandas.DataFrame.transpose(pandas.DataFrame.from_dict(fields, orient='index'))
    df['speechname'] = basename(s) 
    dataframes.append(df)
```

```python
df = pandas.concat(dataframes)
df.set_index('speechname',drop=True)
```
