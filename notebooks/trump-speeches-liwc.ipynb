```python
%matplotlib inline
import requests
import json 
import matplotlib.pyplot as plt
import pandas
from datetime import datetime
import seaborn as sns
import glob
from os.path import basename
sns.set_context('poster')
import re
```

```python
speeches = glob.glob('../results/receptiviti/trump-speeches/*')
```

```python
dataframes = []
for s in speeches:
    j = json.load(open(s))
    fields = {k:v for k,v in j['liwc_scores'].iteritems() if k != 'categories'}
    fields.update(j['liwc_scores']['categories'])
    df = pandas.DataFrame.transpose(pandas.DataFrame.from_dict(fields, orient='index'))
    df['speechname'] = basename(s).replace('.json', '')
    dataframes.append(df)
```

```python
df = pandas.concat(dataframes)
df = df.set_index('speechname',drop=True)
df.to_csv('../results/receptiviti/master/trump-speeches.csv')
```

```python
df.sort_values('tone').tone.plot()
plt.xticks(rotation=50)
plt.ylabel('Negative to Positive')
plt.ylim([0,100])
plt.xlabel('')
plt.title("Emotional tone of Trump's speeches")
plt.savefig('../figs/trump-speeches/trump-tone-ranked.png')
```

```python
df.sort_values('analytic').analytic.plot()
plt.xticks(rotation=50)
plt.ylabel('Analytic score')
plt.ylim([0,100])
plt.xlabel('')
plt.title("Analytic score of Trump's speeches")
plt.savefig('../figs/trump-speeches/trump-analytic-ranked.png')
```

```python
df.sort_values('authentic').authentic.plot()
plt.xticks(rotation=50)
plt.ylabel('Authenticity score')
plt.ylim([0,100])
plt.xlabel('')
plt.title("Authenticity of Trump's speeches")
plt.savefig('../figs/trump-speeches/trump-authentic-ranked.png')
```
